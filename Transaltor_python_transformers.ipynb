```python
from transformers import MarianMTModel, MarianTokenizer
```

```python
#Sample
tokenizer = MarianTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-fr")

text = "When air rises and falls within a thunderstorm, positive and negative charges form in the cloud"
tokenized_input = tokenizer.tokenize(text)
tokenized_input
```

```python
#Implementation
model_name = 'Helsinki-NLP/opus-mt-en-hi'

model = MarianMTModel.from_pretrained(model_name)
tokenizer = MarianTokenizer.from_pretrained(model_name)

def translate_text(input_text):
    input_ids = tokenizer.encode(input_text, return_tensors = "pt")

    translation_ids = model.generate(
        input_ids,
        max_length = 200,
        num_beams = 4,
        length_penalty = 1.0,
        no_repeat_ngram_size = 2,
        top_k = 50,
        top_p = 0.95,
        early_stopping = True
    )

    translated_text = tokenizer.decode(translation_ids[0], skip_special_tokens = True)

    return translated_text

input_text = "Hello, How are you?"
translated_text = translate_text(input_text)
translated_text
```
